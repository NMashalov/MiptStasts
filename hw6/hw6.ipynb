Теория строится на отрицании отрицаний. Можно запутаться :(

Мощность критерия определяется числом ошибок второго рода. Чем таких ошибок меньше, тем критерий мощнее :)

Ошибка второго рода - H_0 неверна, а мы её <s>приняли</s> не отвергли.

Как это померить? Задаём условия, при которых H_0 неверно - истинное значение $\theta_* /ne \theta$ и смотрим, получится ли у критерия <s>принять</s> не опровергнуть альтернативу. Чем больше <s>принятых</s> неопровергнутых альтернатив, тем выше мощность :)


1. Двусторонняя альтернатива:

Альтернатива: $\theta \ne \theta_0$. Истинное значение параметра обозначим за $\theta_*$.

Условия при которых не можем опровергнуть H_1 :

$\beta(\theta_*) := \mathrm{P}(|\frac{\theta -\theta_*}{\sigma}|>z_{\alpha/2})$

Ссылаюсь на https://math.stackexchange.com/questions/1478667/proof-of-the-power-of-the-wald-test

Чем ближе истинное значение к оценке $\theta_0$, тем сложнее их различить. Чем больше выборка, тем больше чувствительность.


Доверительный интервал определяем 

2. Правосторонняя альтернатива - истинное значение правее оценки :)

Альтернатива: $\theta > \theta_0$. Истинное значение параметра обозначим за $\theta_*$.

Условия при которых не можем опровергнуть H1 :

$\beta(\theta_*) := \mathrm{P}(\frac{\theta -\theta_*}{\sigma} > z_{\alpha}) = 1 -\mathrm{P}(\frac{\theta -\theta_*}{\sigma} < z_{\alpha}) $



$$\Phi(z_\alpha -\frac{\theta_0 - \theta_*}{\sigma})$$

3. Левосторонняя альтернатива - истинное значение левее оценки :)
   

Аналогично $$1 - \Phi(z_\alpha + \frac{\theta_0 - \theta_*}{\sigma})$$

Проведен эксперимент, получены данные из экспоненциального распределения. 

```python
sample = [0.11731702, 0.75253036, 0.32918642, 0.22823564, 0.04240622,
        0.04239907, 0.01495969, 0.50280772, 0.22977054, 0.30781252,
        0.00519983, 0.87588937, 0.44660739, 0.05967191, 0.05016975,
        0.05065286, 0.09068843, 0.18598196, 0.14138427, 0.08605575,
        0.23659272, 0.03755863, 0.08637888, 0.1140693 , 0.15223367,
        0.384484  , 0.05568397, 0.18050729, 0.22437618, 0.01189096]
```

Необходимо проверить, является ли это распределение с параметром $\lambda=2$. Используя Критерий Вальда, сделайте вывод по данному предположению.

$X_1, ... X_n$ - выборка из распределения $Exp(\theta)$.  
Проверьте гипотезу $\mathsf{H}_0\colon \theta = 2$ c правосторонней и левосторонней альтернативами.

1. Оформите функцию подсчета статистики критерия, p-value и доверительного интервала 

```python
def wald_test(sample, theta, estimation_theta, estimation_sigma, alternative='two_sided'):
    import numpy as np
    import scipy.stats as sps
    """
    param sample: реализация выборки
    param theta: истинное значение параметра
    param estimation_theta: оценка параметра
    param estimation_sigma: оценка асимптотической дисперсии оценки estimation_sigma
    param alternative: вид альтернативной гипотезы, может принимать одно из значений 'two_sided', 'less', 'greater'

    return statistic
    return p_value
    return conf_int - доверительный интервал
    """

    alpha = 0.05
    n = len(sample)
    statistic = np.sqrt(n) * (estimation_theta - theta) / estimation_sigma

    if alternative == 'two_sided':
        z = sps.norm.ppf(1 - alpha/2)
        p_value = sps.norm.sf(np.abs(statistic)) + sps.norm.cdf(-np.abs(statistic))
        conf_int = np.round(estimation_theta - z*estimation_sigma/np.sqrt(n), 4), np.round(estimation_theta + z*estimation_sigma/np.sqrt(n), 4)


    elif alternative == 'less':
        p_value = sps.norm.cdf(statistic)
        z = sps.norm.ppf(1 - alpha)
        conf_int = (-np.inf, np.round(estimation_theta + z*estimation_sigma/np.sqrt(n), 4))

    
    elif alternative == 'greater':
        p_value = sps.norm.sf(statistic)
        z = sps.norm.ppf(1 - alpha)
        conf_int = (np.round(estimation_theta - z*estimation_sigma/np.sqrt(n), 4), np.inf)

    else:
        raise ValueError('alternative name is wrong')

    return statistic, p_value, conf_int
```

2. Проверьте гипотезу с правосторонней альтернативой.

```python
theta = 2
estimation_theta = 1 / np.mean(sample)
print(estimation_theta)
estimation_sigma = 1 / estimation_theta

wald_test(sample, theta, estimation_theta, estimation_sigma, alternative='greater')
```

**Вывод:** Не опровергаем H1, p value < 0.05.

3. Проверьте гипотезу с левосторонней альтернативой.

```python
print(estimation_theta)
wald_test(sample, theta, estimation_theta, estimation_sigma, alternative='less')
```

**Вывод:** Получаем ответ на вопрос, с какой вероятностью мы отвергаем гипотезу $\theta < 2$. В данном случае отвергаем гипотезу p vlaue > 0.05



### Задача 4

На практике часто рассматривают асимптотические критерии, например, критерий Вальда. Такие критерии контролируют вероятность ошибки I рода на уровне $\alpha$ только в пределе, однако для небольших выборок она может сильно отличаться от $\alpha$, причем не редко в большую сторону. Поэтому для таких критериев возникает необходимость в вычислении значения вероятности ошибки I рода или *реального уровня значимости* при конечной выборке.

Зачастую посчитать точное значение реального уровня значимости довольно сложно, поэтому вместо него вычисляют оценку с помощью семплирования по методу Монте-Карло.

Пусть гипотеза $\mathsf{H}_0$ простая, то есть $\mathsf{H}_0\colon \mathsf{P} = \mathsf{P}_0$. Для ее проверки по конечной выборке $X_1, ..., X_n$ используется критерий $S$. **Опишите схему оценки реального уровня** значимости критерия $S$ методом Монте-Карло.

**Ответ:** 
Реальные уровень значимости можно оценить посчитав сколько раз тест не отверг гипотезу H1, в условиях когда она неверна.

Чтобы создать такие условия генерируем выборку, соответствующую гипотезе H0.

Для A/B тестов есть прекрасная статья: https://habr.com/ru/companies/avito/articles/571094/



Если гипотеза $\mathsf{H}_0$ сложная, то есть $\mathsf{H}_0\colon \mathsf{P} \in \mathscr{P}_0$, то можно повторить такую процедуру несколько раз для разных $\mathsf{P} \in \mathscr{P}_0$.

Рассмотрим альтернативную гипотезу $\mathsf{H}_1\colon \mathsf{P} \in \mathscr{P}_1$. **Опишите схему оценки мощности** критерия $S$ методом Монте-Карло.

**Ответ:** 

Оценку мощности можно провести посчитав сколько раз тест отверг гипотезу H1, в условиях когда она верна.

Чтобы создать такие условия генерируем выборку, соответствующую гипотезе H1.

Пусть $X_1, ..., X_n$ &mdash; выборка из пуассоновского распределения с параметром $\theta$. Рассмотрим гипотезы $\mathsf{H}_0\colon \theta = 1\ \ vs.\ \ \mathsf{H}_1\colon \theta \not= 1$ и критерий Вальда для их проверки.

**1.** **Оцените реальный уровень значимости** критерия для $ n = 50$.

```python
import numpy as np
X = np.random.poisson(lam=1.,size=(10000,50))

theta = 1
estimation_theta = np.mean(X,axis=1)
estimation_sigma = estimation_theta
counter = 0 

for i in range(10000):
    _,p, _= wald_test(X[i], theta, estimation_theta[i], estimation_sigma[i])
    # считаем число не опровергнутых гипотез H1
    counter += p < 0.05

print(counter/10000)
```

**Вывод:** реальный уровень значимости различается с 0.05 на две сотые. Размер выборки удовлетворителен 

**3.** Пусть $n=50$. **Сравните мощности критериев**, соответствующих двусторонней, левосторонней и правосторонней альтернативам при $\theta=0.2$ и при $\theta=2$.

```python
import numpy as np
from itertools import product

n = 5000

x,y = 0.2, 2

for theta in [x,y]:
    X = np.random.poisson(lam=theta,size=(n,50))
    estimation_theta = np.mean(X,axis=1)
    estimation_sigma = estimation_theta


    for hipos in {x : ['two_sided', 'less'],
                  y : ['two_sided', 'greater']}[theta]:
        counter = 0 
        for i in range(n):
            _,p, ci= wald_test(X[i], 1, estimation_theta[i], estimation_sigma[i],alternative=hipos)
            # считаем число не опровергнутых гипотез H1
            counter += p < 0.05
        print(f'theta {theta} alter {hipos} мощность {counter/n}')
```

**Вывод:** мощность признака $\approx 1$ при $\theta_* = 0.2,2$

Статистика T при справедливости H0 имеет t-распределение с n-1 степенью свободы. 

В данном кейсе истинное значение больше оценки -> критерий задаётся как: $T(x) >= t_{n-1,\alpha}$

```python
import scipy.stats as sps
X = sps.norm().rvs(1000)


def my_awesome_ttest(sample,theta=0,alternative='two_sided'):
    """
    param sample: реализация выборки
    param theta: истинное значение параметр
    param alternative: вид альтернативной гипотезы, может принимать одно из значений 'two_sided', 'less', 'greater'

    return statistic
    return p_value
    return conf_int - доверительный интервал
    """
    alpha = 0.05
    df = len(sample) -1
    estimation_theta = np.mean(sample,axis=0)
    estimation_sigma = np.std(sample,axis=0)
    statistic =  df**0.5 * (estimation_theta - theta)/ estimation_sigma 
    
    dist = sps.t(df=df)

    if alternative == 'two_sided':
        z = dist.ppf(1 - alpha/2)
        p_value = dist.sf(np.abs(statistic)) + dist.cdf(-np.abs(statistic))
        conf_int = np.round(estimation_theta - z*estimation_sigma/np.sqrt(df), 4), np.round(estimation_theta + z*estimation_sigma/np.sqrt(df), 4)


    elif alternative == 'less':
        p_value = dist.cdf(statistic)
        z = dist.ppf(1 - alpha)
        conf_int = (-np.inf, np.round(estimation_theta + z*estimation_sigma/np.sqrt(df), 4))

    
    elif alternative == 'greater':
        p_value = dist.sf(statistic)
        z = dist.ppf(alpha)
        conf_int = (np.round(estimation_theta + z*estimation_sigma/np.sqrt(df), 4), np.inf)

    else:
        raise ValueError('alternative name is wrong')
    return statistic, p_value, conf_int


print(sps.ttest_1samp(X,popmean=0, alternative='greater'))

print(my_awesome_ttest(X,alternative='greater'))
```

p_value > 0.05. Отвергаем H1 

2. Пусть  $X = ( X_1, ... X_n)$ &mdash; выборка из неизвестного распределения $\mathsf{P}$ с конечным вторым моментом. Постройте асимптотический критерий для проверки гипотез $\mathsf{H}_0 \colon EX_1 = 0 $   vs.  $\mathsf{H}_1 \colon  EX_1 > 0$ уровня значимости $\alpha$. Покажите, что при $n \to \infty$ этот критерий совпадает с критерием из первого пункта.

ЦПТ: $\sqrt{n} (\overline{X} - \mathrm{E}X) \xrightarrow{d} \mathrm{N}(0,\sigma^2)$

Выборочная дисперсия - состоятельная оценка дисперсии:

$S^2 \xrightarrow{p} \sigma^2$

Используем теорему слуцкого и теорему о наследовании. Переписываем критерий в виде:

$$ \sqrt{n} \frac{\overline{X} - \mathrm{E}X}{S}  \xrightarrow{d} \mathrm{N}(0,1) $$

t-распределение при стремлении $n \rightarrow \infty$ переходит в нормальное. Критерии асимптотически совпадают

```python
import scipy.stats as sps
X = sps.norm().rvs(1000)


def my_asympt_test(sample,theta=0,alternative='two_sided'):
    """
    param sample: реализация выборки
    param theta: истинное значение параметр
    param alternative: вид альтернативной гипотезы, может принимать одно из значений 'two_sided', 'less', 'greater'

    return statistic
    return p_value
    return conf_int - доверительный интервал
    """
    alpha = 0.05
    n = len(sample)
    estimation_theta = np.mean(sample,axis=0)
    estimation_sigma = np.std(sample,axis=0)
    statistic =  n**0.5 * (estimation_theta - theta)/ estimation_sigma 
    
    dist = sps.norm()

    if alternative == 'two_sided':
        z = dist.ppf(1 - alpha/2)
        p_value = dist.sf(np.abs(statistic)) + dist.cdf(-np.abs(statistic))
        conf_int = np.round(estimation_theta - z*estimation_sigma/np.sqrt(n), 4), np.round(estimation_theta + z*estimation_sigma/np.sqrt(n), 4)


    elif alternative == 'less':
        p_value = dist.cdf(statistic)
        z = dist.ppf(1 - alpha)
        conf_int = (-np.inf, np.round(estimation_theta + z*estimation_sigma/np.sqrt(n), 4))

    
    elif alternative == 'greater':
        p_value = dist.sf(statistic)
        z = dist.ppf(alpha)
        conf_int = (np.round(estimation_theta + z*estimation_sigma/np.sqrt(n), 4), np.inf)

    else:
        raise ValueError('alternative name is wrong')
    return statistic, p_value, conf_int


print(my_asympt_test(X,alternative='greater'))
```
