**3.** Реализуйте функцию вычисления выборки оценок $\theta^*_1, \ldots \theta^*_B$ для оценки $\widehat \theta$ методом бутстрепа. Функция должна принимать на вход выборку и количество бутстрепных выборок $B$. 

```python
import numpy as np
from tqdm import tqdm
def estimator(sample,axis=0):
    '''
    Делает необходимые преобразования с выборкой (выполняет подсчет статистики).
    Используется в функции bootstrap для вычисления оценок.
    
    :param sample: массив с числами
    :return: оценка, посчитанная по выборке
    '''
    
    return np.mean(sample,axis=axis)


def bootstrap(sample, B=10**5, batch_size = 10**3):
    '''
    Реализует процедуру бутстрепа.
    
    :param sample: исходная выборка, по которой будут вычисляться бутстрепные оценки
    :param B: число бутстрепных оценок
    :return: возвращает массив длины B c бутстрепными оценками
    '''
    n = len(sample)
    batches = max(B // batch_size,1)

    results = []
    for batch in tqdm(range(batches)):
        indexes= np.random.randint(n, size=(n,batch_size))
        results.append(estimator(sample[indexes]))
    return np.concatenate(results)
```

**4.** Реализуйте три типа бутстрепных доверительных интервалы в виде функций, принимающих на вход выборку оценок, полученных с помощью бутстрепа. Реализовывать вычисление бутстрепных интервалов для подвыборок размера от 1 до n не нужно, только для всей выборки. 

```python
import scipy.stats as sps

def normal_aci(sample, alpha=0.95):
    Z = sps.norm().ppf((1+alpha)/2)
    std = sps.sem(sample)
    sample_estimation = np.mean(sample)
    return (sample_estimation-Z*std,sample_estimation+Z*std)

def boostrap_normal_confidence_interval(bootstrap_estimations, sample_estimation, alpha=0.95):
    Z = sps.norm().ppf((1+alpha)/2)
    v_boot = np.var(bootstrap_estimations)
    return (sample_estimation-Z*v_boot**0.5,sample_estimation+Z*v_boot**0.5)

    
def boostrap_central_confidence_interval(bootstrap_estimations, sample_estimation, alpha=0.95):
    sorted_est = np.sort(bootstrap_estimations)
    B = len(bootstrap_estimations)
    left,right = sorted_est[int(B*(1+alpha)/2)],sorted_est[int(B*(1-alpha)/2)]
    return (2*sample_estimation-left,2*sample_estimation-right)


    
def boostrap_quantile_confidence_interval(bootstrap_estimations, alpha=0.95, **kwargs):
    sorted_est = np.sort(bootstrap_estimations)
    B = len(bootstrap_estimations)
    left,right = sorted_est[int(B*(1-alpha)/2)],sorted_est[int(B*(1+alpha)/2)+1]
    return (left,right)
```

**5.** Для каждого признака постройте бутстрепные доверительные интервалы для $\theta = \mathsf{E} X_1$ и сравните их.

```python
from  functools import partial

round_3 = partial(round,ndigits=3)

ci_funcs = [
    boostrap_normal_confidence_interval,
    boostrap_central_confidence_interval,
    boostrap_quantile_confidence_interval
]

features = ['E1','E2','M']

names = [
    'АНО ДИ',
    'Нормальный ДИ',
    'Центральный ДИ',
    'Квантильный ДИ'
]
   
ci_features = {}   

for feature in features:
    ci_feature = []
    sample = df[feature].dropna().values

    sample_estimation = np.mean(sample)

    ci_feature.append(normal_aci(sample))

    res = bootstrap(sample,B=100_000) 

    for name,ci_func in zip(names,ci_funcs):
        ci_feature.append(ci_func(res,sample_estimation=sample_estimation))
    ci_feature = [tuple(map(round_3,i)) for i in ci_feature]
    ci_features[feature] = ci_feature

print(pd.DataFrame(ci_features,index=names))
```

**Вывод:** Доверительные интервалы бутстрэпа отличаются менее чем на 1%. 

Бутстрэп это долго. "Узкое горлышко" - генерация случайных величин. Разрешают проблему горизонтальным масштабированием вычислительных мощностей. При этом генерация биноминального распределения заменяется на пуассоновское. Опираются на факт $\lim\limits_{n \to \infty} \mathcal{Bin}(n,\frac{1}{n})=Poisson(1)$. При этом нарушаются условие размер выборки может отличаться от n, но при $n > 1000$ это несущественно :smile:

### Задача 5.
На высоте 1 метр от точки $\theta$ находится источник $\gamma$-излучения, который излучает безмассовых $\gamma$-котиков, причем направления траекторий $\gamma$-котиков случайны, т.е. равномерно распределены по полуокружности. Регистрируются координаты $X_i, i=1,...,n$ точек пересечения $\gamma$-котиков с поверхностью детекторной плоскости.
- Найдите оценку максимального правдоподобия параметра $\theta$ для $n=1$ и $n=2$.
- Найдите асимптотическую дисперсию, которую может обещать теорема о свойствах ОМП. Что мешает применить эту теорему? Сравните с оценкой с помощью выборочной медианы.

Угол отклонения частицы от перпендикулярного к плоскости направления $\theta$ связан с $X$ как:

$$x = R tg(\theta)$$

Для связи распределений дифференцируем:


$$dx = R \frac{1}{cos^2\theta} d\theta = R (1+tg^2\theta) d\theta = R (1 + x^2/R^2) d\theta$$


$$ \frac{1}{R} \frac{dx}{1 +x^2/R^2} = d\theta $$

R по условию равен 1 -> x распределён как распределение Коши



Логарифм правдоподобия распределения Коши для выборки размера n: 
$\ell_x(\theta) = -n ln(\pi) - \sum_{i=1}^n ln(1+(x_i-\theta)^2)$

При n=1 $\ell_X' = 2\frac{x_i-\hat{\theta}}{1+(x_1-\hat{\theta})^2} = 0 \rightarrow \hat{\theta} = x_1$

При n=2 $\ell_X' = 2 * (\frac{x_1-\hat{\theta}}{1+(x_1-\hat{\theta})^2} + \frac{x_2-\hat{\theta}}{1+(x_2-\hat{\theta})^2})  = 0 $

Уравнение кубическое :( -> действительных корней может быть более одного -> нарушается условие на единственный максимум для ОМП 



Подробный разбор:

Асимптотическая дисперсия оценки $\sigma^2 = 1 / \ell_X''(\theta)$

$\ell_X'' = \sum_{i=1}^n 2\frac{(\theta-x_i)^2-1}{(1+(x_i-\theta)^2)^2}$

При $\theta = 0$:

$$\sigma^2 = \frac{1}{2* E_x(\frac{(x)^2-1}{(1+(x)^2)^2})} = 2$$

Для выборочной медианы:



$$\sigma^2 = \pi^2/4 = 3.23$$


### Задача 6.


Сгенерируйте выборку $X = \left(X_1, \dots, X_{1000} \right)$ из стандартного распределения Коши. Для всех $n \leqslant 1000$ по первым $n$ элементам выборки $X_1, \dots, X_n$ вычислите значения следующих оценок:
- $\overline{X}$  —  выборочное среднее;
- $\overline{X}_{\alpha}$ —  усеченное среднее, где $\alpha$ &mdash; значение, на котором достигается минимум $\sigma^2_{\alpha}$;
- $\widehat{\mu}$ —  выборочная медиана;

а также, по каждой из этих оценок, одношаговую оценку. 

*Напоминание:* если $\widehat{\theta}_0$ — асимптотически нормальная оценка, то одношаговая оценка $\widehat{\theta}_1$ вычисляется как $\widehat{\theta}_1 = \widehat{\theta}_0 - \left(\ell''_X(\widehat{\theta}_0) \right)^{-1} \ell'_X(\widehat{\theta}_0)$, где $\ell_X(\theta)$ — логарифмическая функция правдоподобия. Заметим, что обычное выборочное среднее не является асимптотически нормальной оценкой, и оценка, вычисленная по формуле выше, формально не является одношаговой, однако ее все равно требуется посчитать.

На одном графике изобразите зависимость значений этих оценок от $n$. Для каждой оценки $\widehat{\theta}_0$ соответствующая оценка $\widehat{\theta}_1$ должна быть изображена на графике пунктиром тем же цветом, что и $\widehat{\theta}_0$. 

Сравните асимптотические дисперсии оценок. 

*Замечание:* если некоторые оценки имеют большой разброс, и разница между графиками зависимостей оценок с малыми значениями недостаточно заметна, стоит сделать два графика, на одном из которых будут изображены все оценки, а на втором &mdash; только достаточно хорошие.


Логарифм правдоподобия:
$\ell_x(\theta) = -n ln(\pi) - \sum_{i=1}^n ln(1+(x_i-\theta)^2)$

Первая производная по логарифму правдоподобия:
$\ell_X' =\sum_{i=1}^n 2\frac{x_i-\theta}{1+(x_i-\theta)^2}$

Вторая производная по логарифму правдоподобия:
$\ell_X'' = \sum_{i=1}^n 2\frac{(\theta-x_i)^2-1}{(1+(x_i-\theta)^2)^2}$


Асимптотическая дисперсия одношаговых оценок:
$\sigma^2 = 1 / \ell_X''(\theta)$
Априорно знаем, что $\theta =0$:
$\sigma^2 = \frac{1}{2* E_x(\frac{(x)^2-1}{(1+(x)^2)^2})}$
Расчёт матожидания:



Асимптотическая оценка дисперсии равна: $\sigma^2 = 2$

Для усеченного среднего и медианы сошлюсь на прошлое задание: $\sigma^2$ = 3.23 и $\frac{\pi^2}{4} \sim 2.46$



```python
import scipy.stats as sps
import matplotlib.pyplot as plt
from  functools import partial

def first_step(sample,theta = 0):
    first_der = 2 * np.sum( (sample - theta)/ (1+(sample-theta)**2))
    second_der = 2 * np.sum( ((sample - theta)**2 -1)/ ((1+(sample-theta)**2))**2)

    return theta - first_der/second_der

start,end = 500,10000 

X = sps.cauchy().rvs(end)

alpha = 0.22229700527367036

means, trim_means, medians = [], [], []

step_means, step_trim_means, step_medians = [], [], []



for i in range(start,end):
    sample = X[:i+1]
    step = partial(first_step,sample=sample)

    means.append(sample.mean(axis=0))
    trim_means.append(sps.trim_mean(sample,alpha,axis=0))
    medians.append(np.median(sample,axis=0))

    step_means.append(step(theta=sample.mean(axis=0)))
    step_trim_means.append(step(theta=sps.trim_mean(sample,alpha,axis=0)))
    step_medians.append(step(theta=np.median(sample,axis=0)))

means, trim_means, medians= np.array(means), np.array(trim_means), np.array(medians)

fig,ax = plt.subplots(1,3,figsize=(15,7))

ax[0].plot(range(start,end),means,label='Оценка выборочным средним')
ax[0].plot(range(start,end),trim_means ,label='Оценка выборочным усеченным средним')
ax[0].plot(range(start,end),medians,label='Оценка выборочной медианой')

ax[0].plot(range(start,end),step_means,label='Оценка одношаговым выборочным средним')
ax[0].plot(range(start,end),step_trim_means ,label='Оценка одношаговым выборочным усеченным средним')
ax[0].plot(range(start,end),step_medians,label='Оценка одношаговой выборочной медианой')


ax[0].set(title = 'Все выборки', xlabel = 'Размер выборки')
ax[0].legend()

ax[1].plot(range(start,end),trim_means,label='Оценка выборочным усеченным средним')
ax[1].plot(range(start,end),medians,label='Оценка выборочной медианой')
ax[1].plot(range(start,end),step_trim_means ,label='Оценка одношаговым выборочным усеченным средним')
ax[1].plot(range(start,end),step_medians,label='Оценка одношаговой выборочной медианой')
ax[1].set(title = 'Выборки без существенного разброса', xlabel = 'Размер выборки')
ax[1].legend()


ax[2].plot(range(start,end),step_medians,label='Оценка одношаговой выборочной медианой')
ax[2].plot(range(start,end),step_trim_means ,label='Оценка одношаговым выборочным усеченным средним')
ax[2].set(title = 'Выборки без существенного разброса', xlabel = 'Размер выборки')
ax[2].legend()
None
```
