Пусть $X_1,...,X_n$ выборка из некоторого распределения $\mathrm{P}$, причем ${\sf D} X_1=\sigma^2<+\infty$, и $\sigma$ неизвестно. 
Рассмотрим статистику $S^2=\frac1n\sum\limits_{i=1}^n(X_i-\overline{X})^2$.

1. Докажите, что статистика $S^2$ равна $\overline{X^2}-\overline{X}^2$.


2. Является ли она состоятельной оценкой $\sigma^2$? Для ответа на этот вопрос воспользуйтесь утверждениями с лекции.


3. Является ли статистика $S^2$ несмещенной оценкой $\sigma^2$? Для ответа на этот вопрос вспомните различные варианты определения дисперсии.

1. Раскрываем квадрат разности $S^2$:
$$S^2 = \frac{1}{n} \left( \sum_{i=1}^n X_i^2 - 2 \cdot \overline{X} \sum_{i=1}^n X_i - \overline{X}^2 \right)$$
Поскольку $\frac{1}{n} \sum_{i=1}^n X_i^2 := \overline{X^2}$, то:

$$S^2 = \overline{X^2} - 2 \cdot \overline{X}^2 + \overline{X}^2 = \overline{X^2} - \overline{X}^2$$

2. Оценка является состоятельной для $\sigma^2$. Действительно,

$\sigma^2 = \mathrm{E}X^2 - (\mathrm{E}X)^2$

По ЗБЧ: 
- $\overline{X^2} \xrightarrow{P} \mathrm{E}X^2 $
- $\overline{X} \xrightarrow{P} \mathrm{E}X$

Тогда $S^2 = \overline{X^2} - \overline{X}^2 \xrightarrow{P} \mathrm{E}X^2 - (\mathrm{E}X)^2$ как разность и произведение состоятельных оценок, сходящихся к константе. 

3. $S^2$ не является несмещенной оценкой дисперсии $\sigma^2$. Частично сошлюсь на [великолепный источник](https://www.statlect.com/fundamentals-of-statistics/variance-estimation#:~:text=Variance%20estimation%20is%20a%20statistical,estimator%20of%20the%20population%20variance.). Key идея: +- истинное матожидание.



Ковариация между независимыми сэмплами = 0 $\rightarrow \mathrm{E}[(X_i-\mu)(X_j-\mu)]=0, i \neq j$. Тогда:

$\mathrm{E}S^2\frac{n+1}{n} \sigma^2 - \frac{2}{n^2} \cdot \sigma^2 = \frac{n-1}{n}\sigma^2$

## Задача 5


Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \sigma^2)$. Известно, что в качестве оценки параметра $\sigma^2$ можно использовать следующие оценки $S^2, \frac{n}{n-1}S^2$. Проверьте эти оценки на несмещенность повторив эксперимент с семинара.

Реализуйте две функции, каждая из которых на вход берет несколько выборок, а на выход выдает массив оценок  
(первая функция - $S^2$, вторая функция - $\frac{n}{n-1}S^2$)  для каждой выборки.  
*Примечание:* У матрицы, заданной в виде `numpy.array` есть метод `var` вычисляющий выборочную дисперсию. 

```python
import numpy as np
def estimate_s2(X):
    '''     
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    # ваш код
    return np.var(X, axis=1)

def estimate_s2_corrected(X):
    '''     
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    # ваш код
    return np.var(X, axis=1,ddof=1)
```

Проверьте правильность своей реализации

```python
X = np.array([[1, 3, 3, 6], [2, 5, 1, 89]])
estimate_s2(X)

assert(sum(estimate_s2(X) != np.array([3.1875, 1399.6875])) == 0)
assert(sum(estimate_s2_corrected(X) != np.array((4.25, 1866.25))) == 0)
```

```python
estimators = [
    (estimate_s2, r'$S^2$', 'purple', 0.1),
    (estimate_s2_corrected, r'$\frac{n}{n-1}S^2$', 'green', 0.2)
]
```

Напишите функцию построения таких графиков в общем виде: на вход функция может получать любые распределения (`distributions`) и любые оценки (`estimators`).




```python
def est_plot(distribution, estimators, sample_size, sample_count, **kargs):
    """
    Построение графика разброса реализаций оценок и их средних значений.

    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    sample_size -- размер выборок
    sample_count -- количество генерируемых выборок
    """
    try:
        X = distribution.rvs(size=(sample_count, sample_size))
    except:
        X = distribution.sample((sample_count, sample_size))
    
    try:
        ax = kargs["ax"]
    except:
        fig, ax = plt.subplots(1, 1, figsize=(8, 8))

    for estimator, name, color, y in estimators:
        E = estimator(X)
        ax.scatter(
            E, np.zeros(sample_count) + y, alpha=0.1, s=100, color=color, label=name
        )
        ax.scatter(E.mean(), y, marker="*", s=300, color="w", edgecolors="black")

        ax.vlines(1, -1, 1, color="r")
        ax.set(
            title="Размер выборки = %d" % sample_size,
            ylim=(0, 0.1 + 0.1 * len(estimators)),
            yticks=[],
        )
        ax.legend()
```

Постройте три графика, аналогичных верхнему, для разных размеров выборок (10, 100, 500).


```python
sample_size_list = (10, 100, 500)  # размеры выборок
sample_count = 500  # количество экспериментов


fig, axes = plt.subplots(3,1,figsize=(15,7))

for ax,sample_size in zip(axes,sample_size_list):
    est_plot(tfd.Normal(loc=0,scale=1), estimators, sample_size, sample_count, ax=ax)
```

Для вычисления зависимости нужно один раз сгенерировать выборки из достаточно большого размера и посчитать оценки по префиксам, используя функции из `numpy`.




```python
# ваш код
import scipy.stats as sps

def mean_plot(distribution, estimators, n_grid, sample_count):
    """
    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    n_grid -- массив значений размера выборки
    sample_count -- количество генерируемых выборок
    """
    sample_size = n_grid.max()

    try:
        X = distribution.rvs(size=(sample_count, sample_size))
    except:
        X = distribution.sample((sample_count, sample_size))

    fig,ax = plt.subplots(1,1,figsize=(8,8))

    for estimator, name, color, y in estimators:
        
        E = []
        for bounder in n_grid:
            E.append(np.mean(estimator(X[:,:bounder])))

        ax.plot(n_grid,E, color=color, label=name)

        ax.legend()

    ax.plot(range(300),np.ones(300), color='red',)



n_grid = np.linspace(10, 300, 100,dtype=int)

mean_plot(sps.norm(loc=0,scale=1), estimators, n_grid, 5000)
```

В этой задаче вам необходимо визуализировать данные свойства.



Зададим параметры эксперимента

```python
theta = 2  # истинное значение параметра
sample_size = 300  # размер выборок
sample_count = 500  # количество выборок
n_range = (np.arange(sample_size) + 1)  # размеры подвыборок
n_range.shape
```

Сгенерируем множество выборок (количество выборок - `sample_count`, размер каждой выборки - `sample_size`)

```python
# генерируем множество выборок, 
# параметр theta является обратным к параметру масштаба
samples = sps.expon(scale=1/theta).rvs(size=(sample_count, sample_size))
samples.shape
```

Вычислим оценки в зависимости от размера выборки, усредняем по последней оси

```python
estimation = n_range / samples.cumsum(axis=1)

print(samples.shape, n_range.shape)
```

Визуализируем для каждой выборки все оценки в зависимости от размера выборки.

Что можно сказать о состоятельности данной оценки?

```python
plt.figure(figsize=(15, 7))
for i in range(sample_count):
    plt.plot(np.arange(sample_size) + 1, estimation[i], color='blue', alpha=0.05)
plt.title('Поведение оценки для разных реализаций')
plt.xlabel('Размер выборки')
plt.ylim((0, 5))
plt.xlim((0, sample_size));
```

## Задача 8
Асимпотические доверительные интервалы ничего не могут гарантировать на малых размерах выборки.
В этой задаче вам предстоит иллюстрировать этот факт, посчитав реальный уровень доверия для интервалов. 

**Реальный уровень доверия**  - доля случаев попадания истинного начения параметра в доверительный интервал для его оценки. 

*Пример:* вы проводите 100 экспериментов (создаете 100 выборок), для каждой выборки рассчитываете АДИ для оценки. Истинный параметр попадает в 93 интервала из 100, значит реальный уровень доверия в этом случае будет 0.93


Сгенерируйте набор выборок из нормального распределения $\mathcal{N}(\theta, 1)$, $\theta=0$

```python

theta = 0  # истинное значение параметра
sample_size = 300  
sample_count = 5000
dist = tfd.Normal(loc=0,scale=1)
X = dist.sample((sample_count,sample_size)).numpy()# ваш код
X.shape
```

Рассмотрим $\hat{\theta} = \overline{X}$.

Напоминание из теории:
$$\hat{\theta} - \frac{z_{\frac{1+\alpha}{2}}\hat{\sigma}}{\sqrt{n}} < \theta < \hat{\theta} + \frac{z_{\frac{1+\alpha}{2}}\hat{\sigma}}{\sqrt{n}}$$

Посичтайте $z$ (используйте функцию `.ppf`)

```python
z = dist.quantile((1+0.95)/2).numpy() # ваш код
print(z)
```

Постройте график зависимости реального уровня доверия от размера выборки. 

```python
# ваш код
n_range = np.arange(sample_size)+1
X_means = X.cumsum(axis=1) / n_range

mask = (X_means > - z / n_range**0.5) & (X_means <  z / n_range**0.5)

plt.plot(n_range,mask.sum(axis=0)/sample_count)
plt.show()
```

Сделайте выводы.

**Вывод:**

Нормальное распределение стабильно - сумма случайных величин из нормальных распределений также распределена нормально. Поэтому асимптотические доверительные интервалы являются доверительными интервалами с n>=1. 

Теоретические выкладки подтверждаются экспериментально. Отклонение реального уровня доверия от теоретического не более 0.025.
